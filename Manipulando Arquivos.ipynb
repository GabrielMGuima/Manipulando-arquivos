# Por que precisamos manipular arquivos?
Os arquivos so essenciais para qualquer tipo de programao, pois fornecem um meio de armazenar e recuperar dados. Atravs da manipulao de arquivos, podemos persistir os dados alm da vida til de um programa especfico.

Para manipular arquivos em Python, primeiro precisamos abrilos. Usamos a funo 'Open()' para isso. Quando terminamos de trabalhar com o arquivo, usamos  a funo 'close()' para liberar recusos.



# Conceito de arquivo em informtica
Um arquivo  um container no compuador onde asinformaes so armazenadas em formato digital. Existem dois tipos de arquivos que podemos manipular em Python: arquivos de texto e arquivos binrios.



# Modos de abertura de arquivo
Existem diferentes modos para abrir um arquivo, como somente leitura ('r'), gravao ('w') e anexar ('a'). O modo de abertura deve ser escolhido de acordo com a operao que iremos realizar no mesmo. 



# Mtodo readline e readlines
O mtodo 'readline()' l uma linha por vez, enquanto 'readlines()' retorna uma lista onde cada elemento  uma linha do arquivo.



# Mtodos write e writelines
Podemos usar 'write()' ou 'writelines()' para escrever em um arquivo. lembre-se, no entanto, de abrir o arquivo no modo correto.



# funes de gerenciamento
Python pode oferece funes para gerenciar arquivos e diretrios. Podemos criar,renomear e excluir arquivos e diretrios usando os mdulos  'os' e 'shutil'



# Tratamento
Tratar erros  uma parte impportante da manipulao de arquivos. Python oferece uma variedade de excees que nos permitem lidar com erros comuns.



# Excees mais comuns
 FileNotFoundError:Lanada quando o arquivo que est sendo aberto no pode ser encontrado no diretrio especificado.

PermissionError:Lanada quando ocorre uma tentativa de abrir um arquivo sem as permisses adequadas para leitura ou gravao.

 IOError:Lanada quando ocorre um erro geral de E/S (entrada/sada) ao trabalhar com o arquivo, como problemas de permisso, falta de espao em disco, entre outros.

 UnicodeDecodeError;Lanada quando ocorre um erro ao tentar decodificar os dados de um arquivo de texto usando uma codificao inadequada.

```python
# Exemplo de readline
Arquivo = open('Texto Para Leitura.txt')
print (Arquivo.readline()) # readline puxa somente a primeira linha
Arquivo.close()
```

```python
# Exemplo de readlines
Arquivo = open('Texto Para Leitura.txt')
print (Arquivo.readlines()) # readlines cria uma lista interravel de todo o conteudo

# while len(linha := Arquivo.readlines()):
#  print(f'{linha}')
Arquivo.close() 
```

```python
#Exemplo de write e writelines

Arquivo= open('Vazil pra estudo.txt')
print(Arquivo.readline())
Arquivo.write("Escrevendo dados em um arquivo.")
Arquivo.writelines(('Escrevendo''um' 'novo''texto'))

Arquivo.close
```

```python
# Exemplos de 'os' e 'shutil'

import os
import shutil

os.mkdir('Exemplo')# Criar Um diretrio 
os.rename('Vazil pra estudo.txt','new.txt')#Renomear um arquivo
os.remove('Exemplo')# Remove um arquivo
shutil.move('Texto Para Leitura.txt','Exemplo') # Move um arquivo

#Obs: No pode ser criados dois arquivos com mesmo nome 
# ao mover um item deve-se atualizar o codigo caso mova denovo
# so se pode renomear um arquivo existente 

```
